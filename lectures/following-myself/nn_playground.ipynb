```python
import torch
```

```python
xenc = torch.randn((4, 4))
W = torch.randn((4, 4), requires_grad=True)
```

```python
losses = []

for i in range(10000):
    # forward pass
    logits = xenc @ W
    counts = logits.exp()
    probs = counts / counts.sum(dim=1, keepdim=True)

    # calculate loss (basically want to get all Ws to 0)
    loss = (W ** 2).mean()
    losses.append(loss.item())
    print(f'loss at iteration {i} is {loss.item()}')
    # print(W.data)

    # backward pass
    W.grad = None
    loss.backward()

    # update
    W.data += -W.grad

W.data
```
